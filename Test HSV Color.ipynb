# Detección de colores mediante HSV

Se puede hacer una detección más facil de colores al cambiar del espacio RGB (Red, Green Blue) al HSV (Hue, Saturation, Value). A continuación se puede apreciar de forma básica como se logra esto.

```python
#Cargamos los componentes necesarios para python y iPython
import matplotlib.pyplot as plt
import numpy as np
import cv2

import ipywidgets as widgets
from ipywidgets import interact, interactive, fixed
from IPython.display import display

from matplotlib.colors import hsv_to_rgb

%matplotlib inline
```

```python
#Función para mostrar la imagenes en forma RGB y no BGR
def plot_rgb(img,fig=None):
    #BGR to RBG just for demostration
    b,g,r = cv2.split(img)
    rgb_img = cv2.merge([r,g,b])
    
    fig,ax = plt.subplots(1)
    ax.imshow(rgb_img),plt.title('RGB')
    #plt.imshow(rgb_img)
    
#Cargamos imágen de manzanas
rgb = cv2.imread("/home/hector/Documents/Test/apples.jpg")
plot_rgb(rgb)
```

A continuación se crean dos mascaras para filtrar los colores "sobre" y "debajo" del colo que deseamos. En la imágen siguiente podemos apreciar como "funciona" el espacio HSV.
<img src="files/HSV.png">

Si ingresamos a http://colorizer.org/ podemos apreciar que H determina que tipo de color queremos, S la saturación, y V el brillo.
<img src="files/HSV2.png">

Si ejecutamos el código (en Jupyter se utiliza shift+intro) se mostraran 6 barras donde podremos ajustar los parametros H,S, y V para cada una de las mascaras y se mostrará en una imágen en color rojo las áreas que pasan el filtro.

Buscamos el color rojo.

```python
lower = np.array([0, 0, 0])
upper = np.array([100, 100, 100])

hsv_rgb = cv2.cvtColor(rgb, cv2.COLOR_BGR2HSV)

mask = cv2.inRange(hsv_rgb, lower, upper)

#fig1,ax1 = plt.subplots(1)
#ax1.imshow(mask),plt.title('RGB')

#plot_rgb(rgb)

#fig,ax = plt.subplots(1)
#ax.imshow(hsv_rgb),plt.title('RGB')
#http://ipywidgets.readthedocs.io/en/latest/examples/Widget%20Basics.html


def hue(h):
    lower[0] = h
    mask = cv2.inRange(hsv_rgb, lower, upper)
    plt.imshow(mask),plt.title('RGB')
def sat(s):
    lower[1] = s
    mask = cv2.inRange(hsv_rgb, lower, upper)
    plt.imshow(mask),plt.title('RGB')
def inte(i):
    lower[2] = i
    mask = cv2.inRange(hsv_rgb, lower, upper)
    plt.imshow(mask),plt.title('RGB')

def hue2(h2):
    upper[0] = h2
    mask = cv2.inRange(hsv_rgb, lower, upper)
    plt.imshow(mask),plt.title('RGB')
def sat2(s2):
    upper[1] = s2
    mask = cv2.inRange(hsv_rgb, lower, upper)
    plt.imshow(mask),plt.title('RGB')
def inte2(i2):
    upper[2] = i2
    mask = cv2.inRange(hsv_rgb, lower, upper)
    plt.imshow(mask),plt.title('RGB')
#interact(hue, x=60)
#widgets.IntSlider(min=0,max=180,step=1,value=0)
interact(hue, h=widgets.IntSlider(min=0,max=180,step=1,value=0))
interact(sat, s=widgets.IntSlider(min=0,max=255,step=1,value=0))
interact(inte, i=widgets.IntSlider(min=0,max=255,step=1,value=0))

interact(hue2, h2=widgets.IntSlider(min=0,max=180,step=1,value=90))
interact(sat2, s2=widgets.IntSlider(min=0,max=255,step=1,value=90))
interact(inte2, i2=widgets.IntSlider(min=0,max=255,step=1,value=90))

#a = widgets.IntSlider()
#display(a)
```

Pasamos los parametros a mano al código que se encuentra abajo para mostrar la imágen con las áreas resultantes

```python
#140,30,180
mask2 = cv2.inRange(hsv_rgb,np.array([110,0,0]),np.array([180,255,255]))
res = cv2.bitwise_and(rgb,rgb, mask= mask2)
plot_rgb(res)
```
